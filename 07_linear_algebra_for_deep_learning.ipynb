# Linear Algebra for Deep Learning

Linear Algebra is the branch of mathematics concerning linear equations and linear functions which are represented through matrices and vectors. It is the language of Deep Learning.

```python
import numpy as np
```

## Vectors and Matrices

A vector is an array of numbers. A matrix is a 2-D array of numbers.

```python
v = np.array([1, 2])
A = np.array([[1, 2], [3, 4]])

print("Vector:\n", v)
print("\nMatrix:\n", A)
```

## Matrix Multiplication

The dot product is a key operation.

```python
product = np.dot(A, v)
print("Matrix-Vector product:\n", product)

# Alternative syntax
print("Using @ operator:\n", A @ v)
```

## Eigenvalues and Eigenvectors

Eigenvectors are vectors that, when transformed by a matrix, only get scaled (not rotated).

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```
