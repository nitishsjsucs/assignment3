# From NumPy to TensorFlow and Keras 3

Keras is a high-level deep learning API that allows for easy and fast prototyping. It is built on top of TensorFlow.

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

## Building a Model

In NumPy, you would manually define weights and matrix multiplications. In Keras, you use Layers.

```python
# NumPy equivalent concept
# Output = Input @ Weights + Bias

# Keras implementation
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(32,)),
    keras.layers.Dense(10, activation='softmax')
])

model.summary()
```

## Training

Keras handles the training loop for you with `.fit()`.

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Dummy data
x_train = np.random.random((1000, 32))
y_train = np.random.random((1000, 10))

model.fit(x_train, y_train, epochs=5, batch_size=32)
```
