# Probability for Deep Learning

Probability theory is the mathematical framework for representing uncertain statements. In deep learning, we use it to handle uncertainty in our models and data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Random Variables and Distributions

We can sample from various probability distributions using NumPy.

```python
# Normal Distribution
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

## Expectation and Variance

Expectation is the mean of a random variable, and variance measures its spread.

```python
print("Mean (Expectation):", np.mean(s))
print("Variance:", np.var(s))
print("Standard Deviation:", np.std(s))
```
