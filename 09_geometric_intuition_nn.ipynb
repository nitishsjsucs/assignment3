# Geometric Intuition of Neural Networks

Neural networks can be thought of as performing geometric transformations on input data to make it linearly separable.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Visualizing Decision Boundaries

Consider a simple classification problem.

```python
# Generate simple data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1]) # AND gate

plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr', s=100)
plt.title("AND Gate Data Points")
plt.xlabel("x1")
plt.ylabel("x2")

# Draw a decision boundary (x1 + x2 - 1.5 = 0)
x_vals = np.linspace(-0.5, 1.5, 10)
y_vals = 1.5 - x_vals
plt.plot(x_vals, y_vals, '--k', label="Decision Boundary")
plt.legend()
plt.grid()
plt.show()
```
