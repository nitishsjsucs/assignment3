# Introduction to TensorFlow Tensors

TensorFlow is an end-to-end open source platform for machine learning. Vectors and matrices in TensorFlow are represented as Tensors.

```python
import tensorflow as tf
print("TensorFlow version:", tf.__version__)
```

## Creating Tensors

TensorFlow has `tf.constant` for immutable tensors and `tf.Variable` for mutable state.

```python
c = tf.constant([[1, 2], [3, 4]])
print("Constant:\n", c)

v = tf.Variable([[1, 2], [3, 4]])
print("\nVariable:\n", v)
```

## Operations

TensorFlow operations support broadcasting similar to NumPy.

```python
print("Addition:", c + 2)
print("Matrix Multiplication:", tf.matmul(c, c))
```
