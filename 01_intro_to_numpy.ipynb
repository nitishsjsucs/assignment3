# Introduction to NumPy

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

```python
import numpy as np
```

## Creating Arrays

We can create NumPy arrays from Python lists or using built-in functions.

```python
# From a list
a = np.array([1, 2, 3])
print("Array a:", a)

# Zeros
b = np.zeros((2, 3))
print("\nZeros b:\n", b)

# Ones
c = np.ones((2, 3))
print("\nOnes c:\n", c)

# Range
d = np.arange(10)
print("\nRange d:", d)
```

## Array Operations and Broadcasting

NumPy allows element-wise operations and broadcasting, which is a powerful mechanism that allows numpy to work with arrays of different shapes when performing arithmetic operations.

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print("Add:", x + y)
print("Multiply:", x * y)

# Broadcasting
z = x * 2
print("Broadcast multiply:", z)
```
